```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/kid.jpg")
```

```python
img_gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray , cmap = "gray")
```

```python

img_gray_rgb = np.stack([img_gray , img_gray , img_gray] , axis = -1)
h , w, c = img_gray_rgb.shape
img_gray_rgb[int(h/2 -60):int(h/2 +60) , int(w/2 -60):int(w/2 +60)] = (0 , 0 ,255)
plt.imshow(cv2.cvtColor(img_gray_rgb , cv2.COLOR_BGR2RGB))
```

```python
img_flip = cv2.flip(img_gray_rgb ,0 )
plt.imshow(cv2.cvtColor(img_flip , cv2.COLOR_BGR2RGB))
```

```python

img_rot = cv2.rotate(img_flip , cv2.ROTATE_90_CLOCKWISE) 
plt.imshow(cv2.cvtColor(img_rot , cv2.COLOR_BGR2RGB))
```
